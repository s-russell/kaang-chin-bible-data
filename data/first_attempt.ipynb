```python
import docx2txt
import re
```

```python
johan_txt = docx2txt.process('Johan.docx')
```

```python
len(johan_txt)
```

```python
johan_lines = [line.strip() for line in johan_txt.split('\n') if len(line.strip()) > 0]
```

```python
def parse_verses_from_chunk(verses_chunk):
    verses = [ v.strip() for v in re.findall('([^\d]+)',verses_chunk)]
    return [{'number': index + 1, 'text': verse} for (index, verse) in enumerate(verses)]
```

```python
johan = []
current_chapter = -1
for line in johan_lines[2:]:
    if line.isnumeric():
        current_chapter = int(line)
    else:
        next_verses = [{ 'chapter': current_chapter, 'number': x['number'], 'text': x['text']} for x in parse_verses_from_chunk(line) ]
        johan += next_verses
```
